```python
from scipy.stats import norm
import numpy as np
```

```python
z, mu, sigma = 125, 100, 25
alpha = norm.cdf(z, loc=mu, scale=sigma)
print(alpha)
```

```python
alpha, mu, sigma = 0.8413, 100, 25
z = norm.ppf(alpha, loc=mu, scale=sigma)
print(z)
```

```python
mu, sigma = 100, 25

alpha = [0.5, 0.6, 0.7, 0.8, 0.9, 0.95, 0.975, 0.99]
inv = norm.ppf(alpha, loc=mu, scale=sigma)
print(inv)
```

```python
inv = [100, 106, 113, 121, 132, 141, 149, 158]
alpha = norm.cdf(inv, loc=mu, scale=sigma)
print(alpha)
```
